```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

prompt = ChatPromptTemplate.from_template(" {topic}")
model = ChatOpenAI(model="gpt-3.5-turbo-1106",openai_api_key=OPENAI_API_KEY)
output_parser = StrOutputParser()

chain = prompt | model | output_parser

chain.invoke({"topic": ""})
```

## Prompt 

prompt  BasePromptTemplate  PromptValue  PromptValue  LLM  ChatModel  BaseMessage 

```python
prompt = ChatPromptTemplate.from_template(" {topic}")

prompt_value = prompt.invoke({"topic": ""})
prompt_value
```

## Model 

 PromptValue  model  model  ChatModel  BaseMessage 

```python
message = model.invoke(prompt_value)
message
```

## Output parser 

 model  output_parser  BaseOutputParser  BaseMessage  StrOutputParser 



```python
output_parser.invoke(message)
```

## 



```python
for chunk in chain.stream({"topic": ""}):
    print(chunk, end="", flush=True)
```
